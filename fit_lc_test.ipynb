```python
def getlsstbandpassobjs():
        """
        General utility to return a list of the baseline LSST bandpasses
        and register them as SNCosmo bandpasses accessible through strings
        like 'LSSTu'.

        Parameters
        ----------
        plot : bool, optional, defaults to False
            plot filter functions obtained
        Returns
        -------
        None

        Examples
        --------
        >>> getlsstbandpassobjs()
        Courtesy of Rahul Biswas

        """
        from astropy.units import Unit
        
        bandPassList = ['lsstu', 'lsstg', 'lsstr', 'lssti', 'lsstz', 'lsstY']
        original_bands = ['u', 'g', 'r', 'i', 'z', 'y']
        banddir = os.path.join(os.getenv('THROUGHPUTS_DIR'), 'baseline')
        
        lsstbands = []
        lsstbp = {}

        for i in range(len(bandPassList)):
            band=bandPassList[i]
            # setup sncosmo bandpasses
            #bandfname = banddir + "/total_" + original_bands[i] + '.dat'
            bandfname = banddir + "/approxLSST_" + original_bands[i] + '_total.dat'

            # register the LSST bands to the SNCosmo registry
            # Not needed for LSST, but useful to compare independent codes
            # Usually the next two lines can be merged,
            # but there is an astropy bug currently which affects only OSX.
            numpyband = pl.loadtxt(bandfname)
            sncosmoband = sncosmo.Bandpass(wave=numpyband[:, 0],
                                               trans=numpyband[:, 1],
                                               wave_unit=Unit('nm'),
                                               name=band)
            sncosmo.registry.register(sncosmoband, force=True)

        return None
```

```python
def registerBands(dirname, prefix=None, suffix=None):
    
    import os
    bands = ['u', 'g', 'r', 'i', 'z', 'y']
    for band in bands:
        fname = os.path.join(dirname, prefix + band + suffix)
        data = pl.loadtxt(fname)
        bp = sncosmo.Bandpass(wave=data[:, 0], trans=data[:, 1], name='approxLSST_'+band)
        sncosmo.registry.register(bp, force=True)
```

```python
def get_des_lightcurve(flname):
        """
        Given a filename, returns a light curve astropy table that conforms with sncosmo requirements
        @param fname The filename of the supernova (relative to data_root)
        """
        fl=open(flname,'r')
        mjd=[]
        flt=[]
        flux=[]
        fluxerr=[]
        z=-9
        z_err=-9
        type=-9
        for line in fl:
            s=line.split()
            if len(s)>0:
                if s[0]=='HOST_GALAXY_PHOTO-Z:':
                    z=(float)(s[1])
                    z_err=(float)(s[3])
                elif s[0]=='OBS:':
                    mjd.append((float)(s[1]))
                    flt.append('des'+s[2])
                    flux.append((float)(s[4]))
                    fluxerr.append((float)(s[5]))
                elif s[0]=='SNTYPE:':
                    type=(int)(s[1])
        
        #Zeropoint
        zp=pl.array([27.5]*len(mjd))
        zpsys=['ab']*len(mjd)
        
        #Make everything arrays
        mjd=pl.array(mjd)
        flt=pl.array(flt, dtype='str')
        flux=pl.array(flux)
        fluxerr=pl.array(fluxerr)
        start_mjd=mjd.min()
        mjd=mjd-start_mjd #We shift the times of the observations to all start at zero. If required, the mjd of the initial observation is stored in the metadata.
        #Note: obviously this will only zero the observations in one filter band, the others have to be zeroed if fitting functions.
        tab = Table([mjd, flt, flux, fluxerr, zp, zpsys], names=('mjd', 'filter', 'flux', 'flux_error', 'zp', 'zpsys'), meta={'name':flname,'z':z, 'z_err':z_err, 'type':type, 'initial_observation_time':start_mjd})
        
        return tab
```

```python
def get_lsst_lightcurve(tab):
        tab_new=tab['MJD', 'FLUXCAL', 'FLUXCALERR', 'FLT']
        tab_new.rename_column('MJD','mjd')
        start_mjd=(tab_new['mjd']).min()
        tab_new['mjd']=tab_new['mjd']-start_mjd
        tab_new.rename_column('FLUXCAL','flux')
        tab_new.rename_column('FLUXCALERR','flux_error')
        tab_new.rename_column('FLT','filter')
        tab_new=Table(tab_new, dtype=['f', 'f', 'f', 'S64'])
        old_filts=['u', 'g', 'r', 'i', 'z', 'Y']
        for f in range(len(old_filts)):
            tab_new['filter'][tab_new['filter']==old_filts[f]]='lsst'+old_filts[f]
        zp=Column(name='zp', data=pl.array([27.5]*len(tab_new['mjd'])))
        zpsys=Column(name='zpsys', data=['ab']*len(tab_new['mjd']))
        tab_new.add_column(zp)
        tab_new.add_column(zpsys)
        
        tab_new.meta={'name':tab.meta['SNID'], 'z':tab.meta['REDSHIFT_FINAL'], 'z_err':tab.meta['REDSHIFT_FINAL_ERR'], 'type':tab.meta['SNTYPE'], 
        'initial_observation_time':start_mjd}
        return tab_new
```

```python
def fit_this(lc):
    #Given a light curve, run fit_lc and return the fitted model
    model=sncosmo.Model('salt2-extended')
    #model.set_source_peakabsmag(-19.3, 'bessellb', 'ab')
    z=lc.meta['z']
    z_err=lc.meta['z_err']
    bounds={'z':(z-3*z_err,z+3*z_err), 'x0':(-3, 3), 'x1':(-5, 5), 'c':(-3, 3)}
    res, fitted_model=sncosmo.fit_lc(lc, model, vparam_names=model.param_names, minsnr=0,bounds=bounds)
    return res, fitted_model
```

Replace the DES filters with LSST filters

```python
des_data_lsst_filters=[]
for d in des_data:
    lc=Table(d.copy(),dtype=['f','S64','f','f','f','S16'])#Create a new table so that astropy table doesn't truncate the strings
    filts=['g','r','i','z']
    for f in filts:
        lc['filter'][lc['filter']=='des'+f]='lsst'+f
        #lc['filter'][lc['filter']==f]='lsst'+f
    des_data_lsst_filters.append(lc)
```

Replace the LSST filters with DES filters (removing the Y band)

```python
lsst_data_des_filters=[]

for d in des_data:
    lc=Table(d.copy(),dtype=['f','S64','f','f','f','S16'])#Create a new table so that astropy table doesn't truncate the strings
    #Remove the Y band
    inds=pl.where(lc['filter']=='lsstY')[0]
    lc.remove_rows(inds)
    filts=['g','r','i','z']
    for f in filts:
        lc['filter'][lc['filter']=='lsst'+f]='des'+f
    lsst_data_des_filters.append(lc)
```

Fit the original DES data

```python
t1=time.time()
des_models=[]
for i in range(len(des_data)):
    res,mod=fit_this(des_data[i])
    des_models.append(mod)
print 'time taken',time.time()-t1 
```

```python
#Check it all worked ok
%matplotlib inline
for i in range(len(des_data)):
    sncosmo.plot_lc(data=des_data[i],model=des_models[i])
```
